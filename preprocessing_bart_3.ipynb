### This notebook is mainly for preprocessing yearly ridership data for each station.

```python
yearly = pd.read_excel("data_source/Daily_Station_Exits.xlsx", sheetname = "Fiscal Year Averages")
```

```python
yearly = yearly[:46]
```

```python
yearly.drop(yearly.columns[-18:], inplace = True, axis = 1)
```

```python
# reformat the dataframe
yearly = pd.melt(yearly, value_vars = yearly.columns[1:], 
        id_vars = ['Station'], var_name = 'Year', value_name = 'num_riders' )
```

```python
yearly['id'] = yearly['Station'].map(lambda(x) : station_mapping[x])
```

```python
yearly.fillna(0, inplace = True)
```

```python
yearly.head()
```

```python
mapping = pd.read_csv("data_source/bart_stations.csv")
final = yearly.merge(mapping, on='id', how='inner', suffixes=('_1', '_2'))
```

```python
final.head()
```

```python
def process_Year(s):
    if 'FY99' in s:
        return 1999
    else:
        return 2000 + int(s[2:])
```

```python
final['Year'] = final['Year'].map(lambda x : process_Year(x))
```
