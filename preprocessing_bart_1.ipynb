```python
import pandas as pd
```

## For Design1

```python
daily = pd.read_excel("data_source/Daily_Station_Exits.xlsx", sheetname = "Data")
```

```python
# drop rows with na for Date
daily = daily[daily.Date.notnull()]
```

```python
daily.drop(daily.columns[-2:], inplace = True, axis = 1)
daily.columns[5:]
```

```python
daily.shape
```

```python
daily = pd.melt(daily, value_vars = daily.columns[5:], 
        id_vars = ['FY', 'Qtr', 'Month', 'Day', 'Date'], var_name = 'id', value_name = 'num_riders' )
```

```python
daily.shape
```

```python
daily.fillna(0, inplace = True)
```

```python
daily['year'] = daily['Date'].dt.year
daily['month'] = daily['Date'].dt.month
daily['weekday'] = daily['Date'].dt.weekday_name
```

```python
len(daily['id'].unique())
```

```python
daily['num_riders'] = [int(x) if x!='CLOSED' and x!= 'SVC DISR' else 0 for x in daily['num_riders']]
```

```python
daily_month_weekday_agg = daily.groupby(['id', 'year', 'month', 'weekday'])['num_riders'].mean().reset_index()
```

```python
daily_month_weekday_agg.shape
```

```python
mapping = pd.read_csv("data_source/bart_stations.csv")
```

```python
len(mapping['id'].unique())
```

```python
final = daily_month_weekday_agg.merge(mapping, on='id', how='inner', suffixes=('_1', '_2'))
```
