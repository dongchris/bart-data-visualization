```python
import pandas as pd
```

```python
daily = pd.read_excel("Daily_Station_Exits.xlsx", sheetname = "Data")
```

```python
daily.drop(daily.columns[-9:], inplace = True, axis = 1)
```

```python
daily.columns[5:]
```

```python
daily.head()
```

```python
daily = pd.melt(daily, value_vars = daily.columns[5:], 
        id_vars = ['FY', 'Qtr', 'Month', 'Day', 'Date'], var_name = 'id', value_name = 'num_riders' )
```

```python
daily.fillna(0, inplace = True)
```

```python
daily['year'] = daily['Date'].dt.year
daily['month'] = daily['Date'].dt.month
daily['weekday'] = daily['Date'].dt.weekday_name
```

```python
daily = daily.iloc[:7394]
```

```python
mapping = pd.read_csv("bart_stations.csv")
```

```python
daily = daily.join(mapping.set_index('id'), how = 'left', on = 'id')
```

```python
daily['num_riders'] = daily['num_riders'].astype(int)
```

```python
daily_month_weekday_agg = daily.groupby(['name', 'year', 'month', 'weekday'])['num_riders'].mean().reset_index()
```
